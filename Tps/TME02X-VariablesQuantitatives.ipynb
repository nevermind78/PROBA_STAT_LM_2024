Une représentation graphique très utile pour comparer facilement des valeurs quantitatives est la <a href="https://fr.wikipedia.org/wiki/Bo%C3%AEte_%C3%A0_moustaches">boîte à moustache</a> : la boîte dénote les données comprises entre le 1er et le 3e quartile ($Q_1$ et $Q_3$), le trait représente la médiane, le triangle la moyenne et les deux segments inférieur et supérieur représente les valeurs $Q_1-1,5*IQR$  et $Q_3+1,5*IQR$, avec $IQR=Q3-Q1$. La ligne suivante permet de tracer les boîtes pour nos données.

<div class="alert alert-block alert-info">
Q3.1 - Déterminer les bornes max et min des longitudes et latitudes dans les variables <code>longmax</code>, <code>longmin</code>, <code>latmax</code>, <code>latmin</code>.
</div>

```python
longmax = np.max(tableau_poi['longitude'])
longmin = np.min(tableau_poi['longitude'])
latmax = np.max(tableau_poi['latitude'])
latmin = np.min(tableau_poi['latitude'])
```

```python
print(longmax,longmin,latmax,latmin, longmax-longmin)
```

```python
c = 1 / (longmax-longmin)
d = 1 / (latmax - latmin) 
```

<div class="alert alert-block alert-info">
Q3.2 - Définir deux nouvelles colonnes <code>longnorm</code> et <code>latnorm</code> qui correspondent aux longitudes et latitudes normalisées entre 0 et 1.
</div>

```python
tableau_poi['longnorm'] = c * (tableau_poi['longitude'] - longmin)
tableau_poi['latnorm'] = d * (tableau_poi['latitude'] - latmin)
tableau_poi
#valeur - valeurmin/ valeurmax - valeurmin
```
